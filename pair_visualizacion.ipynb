```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("/mnt/c/Users/chand/Desktop/ADALAB/Modulo3/Pair_modulo3/Sprint_2/world_data_full_apply_nulos.csv", index_col = 0 ) 
df.head()
```

¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?
Se observa una tendencia descendente

```python
sns.histplot(x = "birth_rate", data = df);
```

¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.
Se observa una tendencia descendente de la esperanza de vida mientras mayor es la mortalidad infantil, son inversamente proporcional
y mayor concentración de esperanza de vida cuanto menor es la mortalidad infantil

```python
sns.scatterplot(x = "infant_mortality", y = "life_expectancy", data = df );
```

¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.
Africa, con el 27.3%

```python
df_continente = df.groupby("continent")["country"].count().reset_index()
```

```python
df_continente.head()
```

```python
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal"]
```

```python
plt.pie("country", labels = "continent", data = df_continente, autopct= '%1.1f%%', colors= colores, textprops={'fontsize': 8}, 
startangle=90); 
```

¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos. 
La mediana es 81. Hay más concentración de datos entre Q1 y Q2 y no se observan datos atípicos

```python
sns.boxplot(x = "life_expectancy", data = df[df["continent"]=="Europe"]);
```
