```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

Archivo CSV sobre el que vamos a generar las visualizaciones

```python
df = pd.read_csv("files/world_data_full_apply_nulos.csv", index_col = 0 ) 
df.head()
```

```python
df.loc[df['birth_rate'].isnull(), :]
```

```python
df = df.dropna(subset=['birth_rate'])
```

 **1 ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?**

```python
df['birth_rate'].sort_values(ascending=True)

#Generar gráfico
plt.figure(figsize=(10, 5))
sns.histplot(x = "birth_rate", data = df,color="violet")
#Visualización
plt.title('Distribución tasa natalidad')
plt.xlabel('Tasa natalidad')
plt.ylabel('Tasa natalidad')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False);
```

 - Interpretación

En el gráfico observamos una relación indirecta en la tasa de natalidad.

Los porcentajes más bajos de natalidad acumulan más países que las tasas de natalidad más altas, donde vemos que el número de países es menor.

**2 ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida?**

```python
plt.figure(figsize=(20, 5))
sns.scatterplot(x = "infant_mortality", y = "life_expectancy", data = df, color= "green", sizes=(20, 200) )
#Visualización
plt.title('Relación entre la tasa de mortalidad infantil y la esperanza de vida')
plt.xlabel('Tasa mortalidad')
plt.ylabel('Esperanza de vida')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False);
```

**3 ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos?**

- Proporciona el porcentaje de países en ese continente

```python
df_continente = df.groupby("continent")["country"].count().reset_index()

total_paises = df_continente['country'].sum()

df_continente['porcentajes'] = np.round((df_continente['country'] / total_paises) * 100,2)


df_continente
```

```python
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal"]

plt.figure(figsize=(30, 10))

plt.pie(df_continente['country'], labels=df_continente['continent'], autopct='%1.1f%%', colors=colores, startangle=90)

#Visualización
plt.title('Distribución de países en %');
```

- Interpretación

El continente con mayor cantidad de países es África, el cual, está formado por 52 países lo que representa un 27.66% del total de países e el conjunto de datos

**5 ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.**

```python
df.head(5)
```

```python
# Crear subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Scatter plot: Tasa de natalidad vs Tasa de mortalidad infantil
sns.scatterplot(data=df, x='birth_rate', y='infant_mortality', ax=axs[0, 0])
axs[0, 0].set_title('Tasa de natalidad vs Tasa de mortalidad infantil')

# Scatter plot: Tasa de natalidad vs Esperanza de vida
sns.scatterplot(data=df, x='birth_rate', y='life_expectancy', ax=axs[0, 1])
axs[0, 1].set_title('Tasa de natalidad vs Esperanza de vida')

# Scatter plot: Tasa de natalidad vs Densidad de población
sns.scatterplot(data=df, x='birth_rate', y='density', ax=axs[1, 0])
axs[1, 0].set_title('Tasa de natalidad vs Densidad de población')

# Scatter plot: Tasa de natalidad vs Población urbana
sns.scatterplot(data=df, x='birth_rate', y='population', ax=axs[1, 1])
axs[1, 1].set_title('Tasa de natalidad vs Población urbana')

# Ajustar el espacio entre subplots
plt.tight_layout()

# Mostrar los subplots
plt.show()
```

- Explicación

No podemos determinar ningún patrón concreto entre estas variables. 

**6 ¿Cuáles son los 10 países con la mayor área forestal?**

```python
df_paises = df[['country', 'forested_area','continent']].sort_values(by='forested_area', ascending=False).head(11)
df_paises.reset_index(drop=True, inplace=True)

palette = sns.color_palette('Greens', n_colors=len(df_paises))

plt.figure(figsize=(20, 6))
plt.barh(df_paises['country'], df_paises['forested_area'], color=palette)
plt.xlabel('Área Forestal (km²)')
plt.ylabel('País')
plt.title('Top 10 Países con Mayor Área Forestal')
plt.gca().invert_yaxis()
plt.show()
```

**8 ¿Cómo se relacionan el CPI y el precio de la gasolina? Crea un scatter plot que muestre esta relación y agrega una línea para visualizar la tendencia. Ajusta los valores del CPI entre 100 y 300.**

```python

df['cpi'] = df['cpi'].clip(lower=100, upper=300)

plt.figure(figsize=(20, 6))
sns.regplot(data=df, x='cpi', y='gasoline_price', scatter_kws={'alpha':0.5})
plt.title('Relación entre CPI y Precio de la Gasolina')
plt.xlabel('Índice de Precios al Consumidor (CPI)')
plt.ylabel('Precio de la Gasolina');
```

- Interpretación

Como podemos ver, la relación entre el CPI y el precio de la gasolina es inversa. A mayor CPI menor precio de fuel.

**10 Queremos analizar la relación entre la tasa de impuestos total, el PIB per cápita y la tasa de alfabetización. Crea un conjunto de subplots con tres gráficas:**
- Un scatter plot que muestre la relación entre el PIB per cápita y el CPI.
- Un histograma de la tasa de impuestos total.
- Un gráfico de barras que resuma la tasa de impuestos total por continente.


```python
df.head()
```

```python
# Crear subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Scatter plot: Relación entre PIB per cápita y CPI
sns.scatterplot(data=df, x='cpi', y='gdp', ax=axs[0])
axs[0].set_title('Relación entre PIB per cápita y CPI')
axs[0].set_xlabel('PIB per cápita')
axs[0].set_ylabel('CPI')

# Histograma de la tasa de impuestos total
sns.histplot(data=df, x='total_tax_rate', bins=10, ax=axs[1], kde=True)
axs[1].set_title('Histograma de la tasa de impuestos total')
axs[1].set_xlabel('Tasa de impuestos total')
axs[1].set_ylabel('Frecuencia')

# Gráfico de barras: Resumen de la tasa de impuestos total por continente
sns.barplot(data=df, x='continent', y='total_tax_rate', ax=axs[2], ci=None)
axs[2].set_title('Resumen de la tasa de impuestos total por continente')
axs[2].set_xlabel('Continente')
axs[2].set_ylabel('Tasa de impuestos total')

# Rotar etiquetas del eje x en el gráfico de barras
axs[2].tick_params(axis='x', rotation=45)

# Ajustar el diseño
plt.tight_layout()

# Mostrar los subplots
plt.show()
```
